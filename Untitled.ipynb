```python
import cv2
import os

# Parameters
num_images = 520
image_prefix = "plot_"
image_suffix = ".jpg"
output_video = 'output3.mp4'
fps = 15
resize_factor = 4  # Factor by which to resize the images

# Initialize list to hold frames
img_array = []

# Initialize size as None
size = None

for i in range(num_images):
    filename = f"{image_prefix}{i}{image_suffix}"
    
    # Check if file exists
    if not os.path.isfile(filename):
        print(f"Warning: {filename} does not exist and will be skipped.")
        continue
    
    # Read the image
    img = cv2.imread(filename)
    
    # Verify if image was read successfully
    if img is None:
        print(f"Warning: {filename} could not be read and will be skipped.")
        continue
    
    # Get original dimensions
    height, width, layers = img.shape
    
    # Calculate resized dimensions
    resized_size = (int(width / resize_factor), int(height / resize_factor))
    print(f"Resizing {filename} to {resized_size}")
    
    # Resize the image
    resized_img = cv2.resize(img, resized_size)
    
    # Set size based on the first successfully read image
    if size is None:
        size = resized_size
        # Initialize VideoWriter after determining size
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # You can try 'XVID' or 'H264' if 'mp4v' doesn't work
        out = cv2.VideoWriter(output_video, fourcc, fps, size)
        if not out.isOpened():
            print("Error: VideoWriter failed to open. Check codec and output path.")
            exit(1)
    
    # Ensure all images have the same size
    if resized_img.shape[1] != size[0] or resized_img.shape[0] != size[1]:
        print(f"Warning: {filename} has a different size after resizing and will be skipped.")
        continue
    
    # Append frame to the video
    out.write(resized_img)
    print(f"Added {filename} to the video.")

# Release the VideoWriter
out.release()
print(f"Video saved as {output_video}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Necessary for 3D plotting

# Example Initialization (Replace with your actual data and parameters)
# X: Array of data points, y: Labels (0 or 1)
X = np.array([[1.0, 1.0], [1.5, 2.0], [2.0, 1.0], [2.5, 1.5], [3.0, 2.0], [3.5, 1.0]])
y = np.array([0, 1, 0, 1, 0, 1])

# Initialize perceptron parameters
weights = [0.5, -0.3]  # [w1, w2]
bias = 0.1
learning_rate = 0.1

# Create meshgrid for plotting
x_min, x_max = 0.9, 2.1
y_min, y_max = 0.9, 2.1
X1, X2 = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))

# Initialize function values and predictions
function_value = np.zeros(len(X))
predicted_outputs = np.zeros(len(X), dtype=int)

# Define datapoints indices
datapoints = range(len(X))

# Set the number of remaining runs
rrs = 2

# Start iterations
iii = 0
while rrs > 0:   
    iii += 1
    w1, w2, b = weights[0], weights[1], bias
    Z = w1 * X1 + w2 * X2 + b
    
    # Update global font size
    plt.rcParams.update({'font.size': 16})
    
    # Create figure with two subplots: Left (3D) and Right (2D)
    fig = plt.figure(figsize=(24, 8), dpi=300)
    
    # --- Left Subplot: 3D Plot ---
    ax1 = fig.add_subplot(1, 2, 1, projection='3d')
    
    # Plot the plane Z = w1*X1 + w2*X2 + b
    plane = ax1.plot_surface(X1, X2, Z, cmap='coolwarm', alpha=0.7)
    fig.colorbar(plane, ax=ax1, shrink=0.5, aspect=10, pad=0.1)
    ax1.set_ylim([0.9, 2.1])
    ax1.set_xlim([0.9, 2.1])
    ax1.set_zlim([-1,1])
    

    
    ax1.set_xlabel('X1')
    ax1.set_ylabel('X2')
    ax1.set_zlabel('Z')
    ax1.set_title('3D Plane (w1, w2, b)')
    
    # --- Right Subplot: 2D Plot ---
    ax2 = fig.add_subplot(1, 2, 2)
    
    # Plot the contour line where Z = 0
    cs = ax2.contour(X1, X2, Z, levels=[0], colors='k', linewidths=2)
    contour_segments = cs.allsegs[0]  # Contour polygons at level=0
    
    # Only proceed if there's at least one contour polygon
    if len(contour_segments) > 0:
        p = contour_segments[0]  # Take the first polygon
        if len(p) > 0:
            mid_idx = len(p) // 2
            px, py = p[mid_idx]  # 'midpoint' in the polygon's vertex list
            
            # Calculate the gradient (normal vector) and normalize it
            grad = np.array([w1, w2], dtype=float)
            norm = np.linalg.norm(grad)
            if norm == 0:
                print("Warning: Gradient vector has zero magnitude.")
                norm_grad = np.array([0, 0])
            else:
                norm_grad = grad / norm  # Normalized gradient vector

            arrow_length = 0.25  # Length of the arrows

            # Calculate arrow end points
            tip_blue = (px - arrow_length * norm_grad[0], py - arrow_length * norm_grad[1])
            tip_red = (px + arrow_length * norm_grad[0], py + arrow_length * norm_grad[1])

            # Arrow pointing toward Z < 0 (Blue)
            ax2.annotate(
                "Predicted Blue",
                xy=tip_blue,          # Arrow tip
                xytext=(px, py),      # Arrow base
                arrowprops=dict(arrowstyle="->", color='blue', linewidth=2),
                color='blue',         # Text color
                fontsize=16,         # Text font size
                ha='center',
                va='bottom',
                bbox=dict(boxstyle="round,pad=0.2", fc="black", alpha=0.7)  # Black background for text
            )

            # Arrow pointing toward Z > 0 (Red)
            ax2.annotate(
                "Predicted Red",
                xy=tip_red,           # Arrow tip
                xytext=(px, py),      # Arrow base
                arrowprops=dict(arrowstyle="->", color='red', linewidth=2),
                color='red',          # Text color
                fontsize=16,          # Text font size
                ha='center',
                va='bottom',
                bbox=dict(boxstyle="round,pad=0.2", fc="black", alpha=0.7)  # Black background for text
            )
    
    # Compute predictions
    for datapoint_index in datapoints:
        datapoint = X[datapoint_index]
        parameter_0 = datapoint[0]
        parameter_1 = datapoint[1]
        function_value[datapoint_index] = parameter_0 * w1 + parameter_1 * w2 + b
        activation = 1 if function_value[datapoint_index] >=0 else 0
        predicted_outputs[datapoint_index] = activation
    
    prediction_error = y - predicted_outputs

    # Debugging and iteration control
    if iii % 10 == 0:
        print(f"Iteration: {iii}")
    pr = np.sum(prediction_error * prediction_error)
    if pr != 0:
        rrs -=1  # Decrement remaining runs if no prediction error
    
    # Update weights and bias
    zeroth_parameter_datapoints = X[:,0]
    weight_0_delta = sum(learning_rate * prediction_error * zeroth_parameter_datapoints)
    first_parameter_datapoints = X[:,1]
    weight_1_delta = sum(learning_rate * prediction_error * first_parameter_datapoints)
    weights[0] += weight_0_delta
    weights[1] += weight_1_delta
    bias += learning_rate * sum(prediction_error)
    
    # Recompute Z with updated weights
    w1, w2, b = weights[0], weights[1], bias
    Z = w1 * X1 + w2 * X2 + b
    
    # Plot decision surface on the 2D subplot
    im = ax2.pcolormesh(X1, X2, Z, cmap='coolwarm', shading='auto', vmin=-0.1, vmax=0.1)
    fig.colorbar(im, ax=ax2)
    
    # Scatter plot of data points
    scatter = ax2.scatter(
        X[:, 0], X[:, 1], c=y, cmap='coolwarm', marker='o',
        edgecolors='black', linewidths=1, s=200  # Increased size with 's=200'
    )
    
    ax2.set_title(f"Iteration {iii}")
    
    # Label each data point as "Correct" or "Incorrect" above the point
    for idx in range(len(X)):
        if predicted_outputs[idx] == y[idx]:
            label_text = "Correct"
            text_color = "lightgreen"
        else:
            label_text = "Incorrect"
            text_color = "lightcoral"

        ax2.annotate(
            label_text,
            (X[idx, 0], X[idx, 1]),
            xytext=(0, 15),  # Increase vertical offset
            textcoords='offset points',
            ha='center',
            color=text_color,
            fontsize=16,  # Adjust font size as needed
            bbox=dict(boxstyle="round,pad=0.2", fc="black", alpha=0.7)  # Black background
        )

    # Set plot limits
    ax2.set_ylim([0.9, 2.1])
    ax2.set_xlim([0.9, 3.5])  # Adjusted to match data range
    
    # Set global title
    fig.suptitle(f"Perceptron Iteration {iii}", fontsize=20)
    
    # Save and close the figure
    #fig.savefig(f"plot_{iii}.jpg", dpi=300)
    #plt.close()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Define plane parameters
w1 = 1    # Coefficient for X-axis
w2 = 2    # Coefficient for Y-axis
b = 3     # Bias term

# Define the range for X and Y
x_min, x_max = -10, 10
y_min, y_max = -10, 10
Z_min, Z_max = -10, 10  # Define the limits for Z (depth)

# Create a meshgrid for plotting the plane
x = np.linspace(x_min, x_max, 10)  # 10 points from -10 to 10 for plane plotting
y = np.linspace(y_min, y_max, 10)
X, Y = np.meshgrid(x, y)
Z = w1 * X + w2 * Y + b

# Initialize the plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
plane = ax.plot_surface(X, Y, Z, alpha=0.5, color='grey', edgecolor='none')

# Define plane corners
corners = [
    [x_min, y_min, w1*x_min + w2*y_min + b],
    [x_max, y_min, w1*x_max + w2*y_min + b],
    [x_max, y_max, w1*x_max + w2*y_max + b],
    [x_min, y_max, w1*x_min + w2*y_max + b]
]

# Define corresponding points at Z_min and Z_max
below_corners = [
    [x_min, y_min, Z_min],
    [x_max, y_min, Z_min],
    [x_max, y_max, Z_min],
    [x_min, y_max, Z_min]
]

above_corners = [
    [x_min, y_min, Z_max],
    [x_max, y_min, Z_max],
    [x_max, y_max, Z_max],
    [x_min, y_max, Z_max]
]

# Create polygons for below the plane
verts_below = [
    [corners[0], corners[1], below_corners[1], below_corners[0]],
    [corners[1], corners[2], below_corners[2], below_corners[1]],
    [corners[2], corners[3], below_corners[3], below_corners[2]],
    [corners[3], corners[0], below_corners[0], below_corners[3]]
]

# Create polygons for above the plane
verts_above = [
    [corners[0], corners[1], above_corners[1], above_corners[0]],
    [corners[1], corners[2], above_corners[2], above_corners[1]],
    [corners[2], corners[3], above_corners[3], above_corners[2]],
    [corners[3], corners[0], above_corners[0], above_corners[3]]
]

# Add Poly3DCollection for below the plane
poly_below = Poly3DCollection(verts_below, facecolors='lightblue', alpha=0.3)
ax.add_collection3d(poly_below)

# Add Poly3DCollection for above the plane
poly_above = Poly3DCollection(verts_above, facecolors='lightcoral', alpha=0.3)
ax.add_collection3d(poly_above)

# Set plot limits
ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)
ax.set_zlim(Z_min, Z_max)

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set title
ax.set_title(f'3D Plane: Z = {w1}X + {w2}Y + {b}')

# Show plot
plt.show()
```
